```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
import time
```

```python
url = "https://ies-midterm.soulution.rocks/login"


r = requests.post(url, data=json.dumps({"cuni": "45167974"})).json()
```

```python
print(r["data"]["personal_code"])
```

```python
list_of_codes = r["data"]["dataset_ids"]
```

```python
ar = []
for i in list_of_codes:
    url = "https://ies-midterm.soulution.rocks/data/" + i +"/"
    ar.append(requests.get(url).json())
```

```python
def return_df(ar):
    data = pd.DataFrame()
    a = ["Open", "High", "Low", "Close", "Adj.Close", "Volume"]
    for comp_data in ar:
        if comp_data["data"]:
            column = comp_data["data"]["company"]
            dictt = {}
            for i in comp_data["data"]["data"]:
                for j in a:
                    if i["Date"] != "":
                        dictt[(i["Date"], j)] = i[j]
        if data.empty:
            data = pd.DataFrame.from_dict(dictt, orient="index")
            data.index = pd.MultiIndex.from_tuples(data.index)
            data.columns = [column]
        else:
            data_p = pd.DataFrame.from_dict(dictt, orient="index")
            data_p.index = pd.MultiIndex.from_tuples(data_p.index)
            data_p.columns = [column]
            data = data.merge(data_p, left_index=True, right_index=True)
    return data
                
    
```

```python
data = return_df(ar)
data.index
```
